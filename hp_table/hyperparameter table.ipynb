```python
# Write to the csv file ('a' means append)
out_file = 'resnet_hp_table.csv'
of_connection = open(out_file, 'a')
writer = csv.writer(of_connection)

for i in range(len(model.layers)):
    class_name = model.layers[i].__class__.__name__
    if class_name=='Conv2D':
        writer.writerow(['Conv2D', 'filters', model.layers[i].filters])
        writer.writerow(['', 'kernel_size', model.layers[i].kernel_size])
        writer.writerow(['', 'strides', model.layers[i].strides])
        writer.writerow(['', 'dilation_rate', model.layers[i].dilation_rate])
        writer.writerow(['', 'groups', model.layers[i].groups])        
    elif class_name=='MaxPooling2D':
        writer.writerow(['MaxPooling2D', 'pool_size', model.layers[i].pool_size])
        writer.writerow(['', 'strides', model.layers[i].strides])
    elif class_name=='BatchNormalization':
        writer.writerow(['BatchNormalization', 'momentum', model.layers[i].momentum])
        writer.writerow(['', 'epsilon', model.layers[i].epsilon])
    elif class_name=='Dense':
        writer.writerow(['Dense', 'units', model.layers[i].units])
    elif class_name=='Dropout':
        writer.writerow(['Dropout', 'rate', model.layers[i].rate])
    elif class_name=='InputLayer':
        writer.writerow(['InputLayer', 'input_shape', model.layers[i].input_shape])
    elif class_name=='ZeroPadding2D':
        writer.writerow(['ZeroPadding2D', 'padding', model.layers[i].padding])
    elif class_name=='SeparableConv2D':
        writer.writerow(['SeparableConv2D', 'filters', model.layers[i].filters])
        writer.writerow(['', 'kernel_size', model.layers[i].kernel_size])
        writer.writerow(['', 'strides', model.layers[i].strides])
        writer.writerow(['', 'dilation_rate', model.layers[i].dilation_rate])
        writer.writerow(['', 'depth_multiplier', model.layers[i].depth_multiplier])
    elif class_name=='AveragePooling2D':
        writer.writerow(['AveragePooling2D', 'pool_size', model.layers[i].pool_size])
        writer.writerow(['', 'strides', model.layers[i].strides])
    elif class_name=='ReLU':
        writer.writerow(['ReLU', 'max_value', model.layers[i].max_value])
        writer.writerow(['', 'negative_slope', model.layers[i].negative_slope])
        writer.writerow(['', 'threshold', model.layers[i].threshold])
    elif class_name=='DepthwiseConv2D':
        writer.writerow(['DepthwiseConv2D', 'kernel_size', model.layers[i].kernel_size])
        writer.writerow(['', 'strides', model.layers[i].strides])
        writer.writerow(['', 'depth_multiplier', model.layers[i].depth_multiplier])
        writer.writerow(['', 'dilation_rate', model.layers[i].dilation_rate])
    elif class_name=='Reshape':
        writer.writerow(['Reshape', 'target_shape', model.layers[i].target_shape])
    elif class_name=='Concatenate':
        writer.writerow(['Concatenate', 'axis', model.layers[i].axis])
        
of_connection.close()
```
